```python
def plot_map(df, column, title):
    fig = px.choropleth(
        df,
        locations = 'Entity',
        locationmode = 'country names',
        color = column,
        hover_name = 'Entity',
        color_continuous_scale = 'RdYlGn',
        animation_frame = 'Year',
        range_color = [0, 100])

    fig.update_geos(
        showcoastlines = True,
        coastlinecolor = "Black",
        showland = True,
        landcolor = "white",
        showcountries = True,
        showocean = True,
        oceancolor = "LightBlue")

    fig.update_layout(
        title_text = title,
        geo = dict(
            showframe = False,
            showcoastlines = False,
            projection_type = 'equirectangular',
            showland = True,
            landcolor = "white",
            showcountries = True,
            showocean = True,
            oceancolor = "LightBlue"),
        width = 1000,
        height = 850,
        dragmode = 'pan',
        hovermode = 'closest',
        coloraxis_colorbar = dict(
            title = column,
            title_font_size = 14,
            title_side = 'right',
            lenmode = 'pixels',
            len = 300,
            thicknessmode = 'pixels',
            thickness = 15),
        updatemenus = [
            {"type": "buttons", "showactive": False, "x": 0.1, "y": 0.9, "buttons": [{"label": "Play", "method": "animate"}]},
            {"type": "buttons", "showactive": False, "x": 0.18, "y": 0.9, "buttons": [{"label": "Pause", "method": "animate"}]},
            {"type": "buttons", "showactive": False, "x": 0.26, "y": 0.9, "buttons": [{"label": "Stop", "method": "animate"}]}],
        sliders = [{"yanchor": "top", "xanchor": "left", "currentvalue": {"font": {"size": 20}}, "steps": []}])
    slider_steps = []

    for year in df['Year'].unique():
        step = {
            "args": [
                [year],
                {"frame": {"duration": 300, "redraw": False}, "mode": "immediate", "transition": {"duration": 0}}],
            "label": str(year),
            "method": "animate"}
        slider_steps.append(step)

    fig.layout.updatemenus[0].buttons[0].args[1]['steps'] = slider_steps

    return fig
```

```python
plot_map(df_analise, 'Access to electricity (% of population)', 'Acesso a eletricidade (% População) ao longo dos anos')
```

```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
features = df.drop(['Access to electricity (% of population)','Entity','Year','Financial flows to developing countries (US $)',], axis=1)
target = df['Access to electricity (% of population)']

xtrain, xtest, ytrain, ytest = train_test_split(features, target, test_size=0.2, random_state=40)

xtrain.shape, xtest.shape

norm = MinMaxScaler()
xtrain = norm.fit_transform(xtrain)
xtest = norm.transform(xtest)
```

```python
# Crie uma instância do modelo LinearRegression
linear_model = LinearRegression()

# Treine o modelo
linear_model.fit(xtrain, ytrain)

# Faça previsões usando o modelo treinado
y_train_pred = linear_model.predict(xtrain)
y_test_pred = linear_model.predict(xtest)

# Calcule as métricas de erro
mse_train = mean_squared_error(ytrain, y_train_pred)
mae_train = mean_absolute_error(ytrain, y_train_pred)
r2_train = r2_score(ytrain, y_train_pred)

mse_test = mean_squared_error(ytest, y_test_pred)
mae_test = mean_absolute_error(ytest, y_test_pred)
r2_test = r2_score(ytest, y_test_pred)

print('LinearRegression:')
print('Training MSE:', mse_train)
print('Validation MSE:', mse_test)
print('Training MAE:', mae_train)
print('Validation MAE:', mae_test)
print('Training R²:', r2_train)
print('Validation R²:', r2_test)
print()
```

```python
rf_model = RandomForestRegressor()
rf_model.fit(xtrain, ytrain)

y_train_pred = rf_model.predict(xtrain)
y_test_pred = rf_model.predict(xtest)

mse_train = mean_squared_error(ytrain, y_train_pred)
mae_train = mean_absolute_error(ytrain, y_train_pred)
r2_train = r2_score(ytrain, y_train_pred)

mse_test = mean_squared_error(ytest, y_test_pred)
mae_test = mean_absolute_error(ytest, y_test_pred)
r2_test = r2_score(ytest, y_test_pred)

print('RandomForestRegressor:')
print('Training MSE:', mse_train)
print('Validation MSE:', mse_test)
print('Training MAE:', mae_train)
print('Validation MAE:', mae_test)
print('Training R²:', r2_train)
print('Validation R²:', r2_test)
print()
```

```python
xgb_model = XGBRegressor()

# Treine o modelo
xgb_model.fit(xtrain, ytrain)

# Faça previsões usando o modelo treinado
y_train_pred = xgb_model.predict(xtrain)
y_test_pred = xgb_model.predict(xtest)

# Calcule as métricas de erro
mse_train = mean_squared_error(ytrain, y_train_pred)
mae_train = mean_absolute_error(ytrain, y_train_pred)
r2_train = r2_score(ytrain, y_train_pred)

mse_test = mean_squared_error(ytest, y_test_pred)
mae_test = mean_absolute_error(ytest, y_test_pred)
r2_test = r2_score(ytest, y_test_pred)

print('XGBRegressor:')
print('Training MSE:', mse_train)
print('Validation MSE:', mse_test)
print('Training MAE:', mae_train)
print('Validation MAE:', mae_test)
print('Training R²:', r2_train)
print('Validation R²:', r2_test)
print()
```

```python
linear_r2 = linear_model.score(xtest, ytest)
rf_r2 = rf_model.score(xtest, ytest)
xgb_r2 = xgb_model.score(xtest, ytest)

# Imprima os R²
print('LinearRegression R²:', linear_r2)
print('RandomForestRegressor R²:', rf_r2)
print('XGBRegressor R²:', xgb_r2)
```

```python
import matplotlib.pyplot as plt

# Previsões dos modelos nos dados de teste
y_linear_pred = linear_model.predict(xtest)
y_rf_pred = rf_model.predict(xtest)
y_xgb_pred = xgb_model.predict(xtest)

# Gráfico de dispersão para Regressão Linear
plt.figure(figsize=(8, 6))
plt.scatter(ytest, y_linear_pred, color='blue', label='Previsões', alpha=0.5)
plt.scatter(ytest, ytest, color='red', label='Valores Reais', alpha=0.5)
plt.xlabel('Valor Real')
plt.ylabel('Previsão')
plt.title('Regressão Linear - Valor Real vs. Previsão')
plt.legend()
plt.grid(True)
plt.show()

# Gráfico de dispersão para Random Forest
plt.figure(figsize=(8, 6))
plt.scatter(ytest, y_rf_pred, color='green', label='Previsões', alpha=0.5)
plt.scatter(ytest, ytest, color='red', label='Valores Reais', alpha=0.5)
plt.xlabel('Valor Real')
plt.ylabel('Previsão')
plt.title('Random Forest - Valor Real vs. Previsão')
plt.legend()
plt.grid(True)
plt.show()

# Gráfico de dispersão para XGBoost
plt.figure(figsize=(8, 6))
plt.scatter(ytest, y_xgb_pred, color='purple', label='Previsões', alpha=0.5)
plt.scatter(ytest, ytest, color='red', label='Valores Reais', alpha=0.5)
plt.xlabel('Valor Real')
plt.ylabel('Previsão')
plt.title('XGBoost - Valor Real vs. Previsão')
plt.legend()
plt.grid(True)
plt.show()
```
